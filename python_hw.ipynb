```python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
list_1 = []
list_2 = []
for i in lst:
    if i == 'robot':
        list_1.append('0')
    elif i == 'human':
        list_1.append('1')
for i in lst:
    if i == 'human':
        list_2.append('0')
    elif i == 'robot':
        list_2.append('1')
```

```python
df = pd.DataFrame({'human':list_1, 'robot':list_2})
df.head(20)
```
